```python
#Standard Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Get the data ready 
heart_disease = pd.read_csv("heart.csv")
heart_disease.head()
```

```python
X = heart_disease.drop('target',axis=1)
# create X (features matrix)
X
```

```python
# Create Y(label matrix / labels)
Y = heart_disease['target']
Y
```

```python
# To deal with warnings 
import warnings
warnings.filterwarnings("ignore") 
# a band-aid solution for warning pop-ups
#“Do not display any warning messages.”
#Sometimes libraries (like pandas or sklearn) show many warnings that clutter the output.
```

```python
# 2.Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# we ll keep the default hyperparameters 
clf.get_params()
# Parameters = learned by the model
# Hyperparameters = chosen by you
#Hyperparameters are settings you choose before training that control how a model learns.
```

```python
# 3.Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
clf.fit(X_train, Y_train);   #.fit-> train the model 
# the model studies patterns in x_train and tries to understand how to relate to y_train
```

```python
y_preds =  clf.predict(X_test)
y_preds
# model guesses the answers for new data
```

```python
Y_test
```

```python
# 4. Evaluate the model on the training data and test data 
clf.score(X_train, Y_train)
# how well did the model learn the training data
```

```python
clf.score(X_test, Y_test)
# This calculates accuracy on test data 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test,y_preds))
```

```python
confusion_matrix(Y_test, y_preds)
```

```python
accuracy_score(Y_test, y_preds)
```

```python
# 5. improve model accuracy 
# try the model accuracy with different n_estimators 
np.random.seed(14) # ensures the results are reproducible 
for i in range(10, 200, 10):
  print(f'Trying model with {i} estimators')
  clf = RandomForestClassifier(n_estimators = i).fit(X_train, Y_train)
  print(f'Model accuracy on the test set: {clf.score(X_test, Y_test)* 100:.2f}%')
    # .score() = predict + calculate accuracy
```

```python
# 6. Save a model amd load it 
import pickle 
# pickle is a Python library used to save objects to a file.
pickle.dump(clf, open('random_forest_model_1.pkl','wb'))
#This line:
#Takes your trained model clf
#Converts it into a special binary format
#Saves it into a file called: random_forest_model_1.pkl
#w → write; b → binary mode
#We must use binary mode because pickle stores data in binary format.
# We save the model so we don’t have to retrain it every time we run the program.
# A pickle (.pkl) file is a file that stores a trained Python object (like a model) in binary form so it can be loaded later.
# We use binary form because it allows Python to store the full complex model object exactly as it exists in memory.
```

```python
loaded_model = pickle.load(open('random_forest_model_1.pkl','rb'))
loaded_model.score(X_test,Y_test)
```

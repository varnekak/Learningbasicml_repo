```python
#Standard Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Get the data ready 
heart_disease = pd.read_csv("heart.csv")
heart_disease.head()
```

```python
X = heart_disease.drop('target',axis=1)
# create X (features matrix)
X
```

```python
# Create Y(label matrix / labels)
Y = heart_disease['target']
Y
```

```python
# To deal with warnings 
import warnings
warnings.filterwarnings("ignore") 
# a band-aid solution for warning pop-ups
#“Do not display any warning messages.”
#Sometimes libraries (like pandas or sklearn) show many warnings that clutter the output.
```

```python
# 2.Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# we ll keep the default hyperparameters 
clf.get_params()
# Parameters = learned by the model
# Hyperparameters = chosen by you
#Hyperparameters are settings you choose before training that control how a model learns.
```

```python
# 3.Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
clf.fit(X_train, Y_train);
```

```python
y_preds =  clf.predict(X_test)
y_preds
```

```python
Y_test
```

```python
# 4. Evaluate the model on the training data and test data 
clf.score(X_train, Y_train)
```

```python
clf.score(X_test, Y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test,y_preds))
```

```python
confusion_matrix(Y_test, y_preds)
```

```python
accuracy_score(Y_test, y_preds)
```

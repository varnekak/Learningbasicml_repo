# Linear regression in pure Numpy (mini Ml example)

```python
#data 
X = np.array([[1,1],[1,2],[1,3]]) # bias + feature 
y = np.array([2,3,4])

#weights 
w = np.array([0.5,1.0])

#prediction
y_hat = X@ w 

# loss(MSE )
loss = ((y - y_hat)**2).mean()

print(y_hat)
print(loss)
```

If your data is in meters:
Variance is in meters²
That’s hard to interpret

Taking square root:
Brings it back to meters
Makes it interpretable in original units

*low variance* 
Points are close to the mean
Small spread
Small variance
Small standard deviation

*High variance*
Points are far from the mean
Large spread
Large variance
Large standard deviation

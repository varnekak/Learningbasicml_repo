# An end-to-end Scikit-learn worklow

```python
import pandas as pd
import numpy as np
```

```python
# 1. Get the data ready 
heart_disease = pd.read_csv("heart.csv")
heart_disease.head()
```

```python
X = heart_disease.drop('target',axis=1)
# create X (features matrix)
X
```

```python
# Create Y(label matrix / labels)
Y = heart_disease['target']
Y
```

```python
# To deal with warnings 
import warnings
warnings.filterwarnings("ignore") 
# a band-aid solution for warning pop-ups
#“Do not display any warning messages.”
#Sometimes libraries (like pandas or sklearn) show many warnings that clutter the output.
```

```python
# 2.Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# we ll keep the default hyperparameters 
clf.get_params()
# Parameters = learned by the model
# Hyperparameters = chosen by you
#Hyperparameters are settings you choose before training that control how a model learns.
```

```python
# 3.Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
clf.fit(X_train, Y_train);   #.fit-> train the model 
# the model studies patterns in x_train and tries to understand how to relate to y_train
```

```python
y_preds =  clf.predict(X_test)
y_preds
# model guesses the answers for new data
```

```python
Y_test
```

```python
# 4. Evaluate the model on the training data and test data 
clf.score(X_train, Y_train)
# how well did the model learn the training data
```

```python
clf.score(X_test, Y_test)
# This calculates accuracy on test data 
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test,y_preds))
```

```python
confusion_matrix(Y_test, y_preds)
```

```python
accuracy_score(Y_test, y_preds)
```

```python
# 5. improve model accuracy 
# try the model accuracy with different n_estimators 
np.random.seed(14) # ensures the results are reproducible 
for i in range(10, 200, 10):
  print(f'Trying model with {i} estimators')
  clf = RandomForestClassifier(n_estimators = i).fit(X_train, Y_train)
  print(f'Model accuracy on the test set: {clf.score(X_test, Y_test)* 100:.2f}%')
    # .score() = predict + calculate accuracy
```

```python
# 6. Save a model amd load it 
import pickle 
# pickle is a Python library used to save objects to a file.
pickle.dump(clf, open('random_forest_model_1.pkl','wb'))
#This line:
#Takes your trained model clf
#Converts it into a special binary format
#Saves it into a file called: random_forest_model_1.pkl
#w → write; b → binary mode
#We must use binary mode because pickle stores data in binary format.
# We save the model so we don’t have to retrain it every time we run the program.
# A pickle (.pkl) file is a file that stores a trained Python object (like a model) in binary form so it can be loaded later.
# We use binary form because it allows Python to store the full complex model object exactly as it exists in memory.
```

```python
loaded_model = pickle.load(open('random_forest_model_1.pkl','rb'))
loaded_model.score(X_test,Y_test)
```

## Examples of Preprocessing Tasks

### 1. Cleaning Data
- Removing symbols like `$` or `,`
- Fixing incorrect values

### 2. Handling Missing Values
- Filling missing data with mean/median
- Removing rows

### 3. Encoding Categorical Data
- Converting text like `"Red"` or `"Toyota"` into numbers
- Example: using `OneHotEncoder`

### 4. Scaling Numeric Data
- Making values comparable (e.g., `0–1` range)

### 5. Feature Selection or Transformation
- Choosing useful columns for the model

```python
# Turn the categories into number 
from sklearn.preprocessing import OneHotEncoder 
#Preprocessing is the step where you clean and transform raw data into a format that a machine learning model can understand and use effectively.
#OneHotEncoder converts categorical values into multiple binary (0/1) columns so machine learning models can use them.
from sklearn.compose import ColumnTransformer 
#ColumnTransformer allows you to apply different preprocessing steps to different columns in a dataset.
#ColumnTransformer from scikit-learn is a preprocessing utility that allows you to apply different transformations to different columns of a dataset at the same time.
# Categorical columns → Encoding (e.g., OneHotEncoder)
#Numeric columns → Scaling (e.g., StandardScaler)
#One-Hot Encoding: Converts each category into separate binary columns (0 or 1).
#Label Encoding: Converts categories into integer labels (0,1,2…).
```

```python
car_sales
```

```python
categorical_features =['Make','Colour','Doors'] # You are creating a list of columns that contain categorical data.
one_hot = OneHotEncoder() #You are creating a list of columns that contain categorical data.

transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)], remainder='passthrough')
# create a columTransformer to apply apply OneEncoder to the categorial columns
# syntax = ('name_of_transform', transformation_method, columns_to_apply)

transformed_X = transformer.fit_transform(X)
# Fit the transformer to the dataset and transform the categorical columns 

transformed_X
# display the transfered dataset after encoding 
```

```python
pd.DataFrame(transformed_X)
```

```python
# another way to do this is with pandas dummies 
dummies = pd.get_dummies(car_sales[['Make','Colour','Doors']])
dummies 
```

```python
# fit the model 
np.random.seed(42)
X_train,X_test, Y_train, Y_test = train_test_split(transformed_X, Y, test_size=0.2)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
```

### 1.2 What if there was missing value?
1. Fill the missing vakues with some value
2. Remove the samples with the missing data altogether



```python
car_sales_missing = pd.read_csv('car-sales-extended-missing-data.csv')
car_sales_missing
```

```python
car_sales_missing.isna().sum()
#It tells you how many missing values (NaN) are present in each column of the dataset.
```

```python
# Fill the make column 
car_sales_missing['Make'].fillna('missing', inplace=True)
car_sales_missing['Colour'].fillna('missing', inplace = True)
car_sales_missing['Doors'].fillna(4, inplace=True)
car_sales_missing['Odometer (KM)'].fillna(car_sales_missing['Odometer (KM)'].mean(),inplace = True)
    
```

```python
car_sales_missing.isna().sum()
```

```python
# remove rows with missing values 
car_sales_missing.dropna(inplace = True)
```

```python
car_sales_missing.isna().sum()
```

```python
X = car_sales_missing.drop('Price', axis =1)
Y = car_sales_missing['Price']
# convert the categories inot numerical values 
categorical_features = ['Make', 'Colour','Doors']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)], remainder ='passthrough')
transformed_X = transformer.fit_transform(X)
X_train, X_test , Y_train, Y_test = train_test_split(transformed_X,Y,test_size = 0.2)
```

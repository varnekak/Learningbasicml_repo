```python
# converting price object to integer 
# car_sales['Price'] = car_sales['Price'].replace({'//$':'',',':''}, regex = True).astype(float).astype(int)
car_sales['Price'] = car_sales["Price"].astype(str).str.replace('[\\$,]','',regex = True).astype(float).astype(int)
car_sales['Price']
```

astype(str)
Converts the Price column to string type. This is necessary because str.replace works on strings.

str.replace('[\\$,]','', regex=True)
Removes dollar signs $ and commas , from the string. The regex=True allows using the pattern [\\$,] as a regular expression.

astype(float)
Converts the cleaned string into a float. This is important because some prices might be decimals.

astype(int)
Converts the float into an integer, giving a clean numeric representation of the price.

Assignment back to car_sales['Price']
Replaces the original Price column with the cleaned integer values.

# Plotting multiple lines 


```python
y2 = [5, 15, 20, 35]
plt.plot(x,y, label = "line 1")
plt.plot(x,y2, label="line 2")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Multiple Lines")
plt.legend() # uses the label values 
plt.show()
```

 # Common Plot Types

**Scatter Plot**

```python
plt.scatter(x,y) # for scatter plot 
plt.title("Scatter Plot")
plt.show()
#good to see relationships 
```

**Bar graph**

```python
categories = ['A','B','C']
val = [19,50,67]
plt.bar(categories, val)
plt.title("Bar graph")
plt.show()
```

**Histogram**

```python
import random 
data = [random.randint(1,100) for _ in range(1000)] #creating a list called data 
#generates one random integer min1 to max 100 both ends included 
plt.hist(data, bins =20)
plt.title("Histogram")
plt.show()
```

# The Object-Oriented Way

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title(" OO Style Plot ")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
# 1st method
fig = plt.figure() #creates a figure 
ax = fig.add_subplot() # adds some axes 
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)# adding some data 
plt.show()
```

```python
#3rd method 
fig, ax = plt.subplots()
ax.plot(x,[50,100,200,250])
plt.show()
```

# For subplots 

```python
fig, ((ax1,ax2),(ax3,ax4))= plt.subplots( nrows = 2 , ncols =2, figsize =(10,5))# creates 4 graph boxes 
#ax1, ax2, ax3, ax4 are four separate plotting areas, arranged in a 22 grid.
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10)) 
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(x)
plt.show()
```

```python
fig, ax= plt.subplots( nrows = 2 , ncols =2, figsize =(10,5))# creates 4 graph boxes 
#ax1, ax2, ax3, ax4 are four separate plotting areas, arranged in a 22 grid.
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10)) 
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(x)
plt.show()
```

# which one should you use?
(pyplot vs matplotlib OO method?)
when plotting something quickly okay to use the pyplot method
when plotting something more advanced, use the OO method 


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# Pyplot method 
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target')
plt.show()
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,4))
over_50.plot(
    kind = 'scatter',x='age',y='chol', c ='target', ax=ax)
plt.show()
```

```python
# OO method from scratch 
fig,ax = plt.subplots(figsize=(10,6)) 

# Plot the data 
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c = over_50['target'])

#customize the plot 
ax.set (title = 'Heart Disease and Cholestrol Levels',
        xlabel =  'Age',
        ylabel = 'Cholestrol')

# Add a legend 
ax.legend(*scatter.legend_elements(), title = 'Target')

#  add a hline 
ax.axhline(over_50['chol'].mean(),linestyle='--')# displays the mean line
plt.show()

```

```python
# Subplots of age, thalach, chol
fig, (ax1, ax2) = plt.subplots( nrows=2, ncols= 1,figsize=(10,10), sharex=True)# shares x axis with the other plots 
# add the data to ax1
scatter = ax1.scatter( x = over_50['age'],
                      y = over_50['chol'],
                      c= over_50['target'])
#customize ax1
ax1.set(title =  'Heart disease and cholestrol levels',ylabel = 'Cholestrol')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['chol'].mean(), linestyle ='--')

#add the data to ax2
scatter = ax2.scatter( x = over_50['age'],
                       y = over_50['thalach'],
                       c= over_50['target'])
#customize ax2
ax2.set(title="Heart disease and Heart beat", ylabel='thalach')
ax2.legend(*scatter.legend_elements(), title= 'Target')
ax2.axhline(y=over_50['thalach'].mean(), linestyle='--')

#add title to the figure 
fig.suptitle('Heart Disease Analysis ', fontsize=16, fontweight='bold')
plt.show()


        
```

```python
# customizing x and y limitations 
plt.style.use('dark_background')
# subplots of age, thalach, chol
fig,(ax0, ax1) = plt.subplots(nrows = 2, ncols= 1, figsize =(10,10), sharex = True) # Share the x axis with other plots 
#  add data to ax0
scatter = ax0.scatter( x = over_50['age'],
                       y = over_50['chol'],
                       c = over_50['target'],
                       cmap= 'winter')
# customize ax0
ax0.set( title = 'Heart Disease and cholestrol levels', ylabel = 'Cholestrol')
ax0.legend(*scatter.legend_elements(), title = 'Target')
ax0.axhline( y = over_50['chol'].mean(),
            linestyle='--')

# setting axis limits 
ax0.set_xlim([50,80])
ax0.set_ylim([100,600])

# Add the data to ax1
scatter = ax1.scatter( x = over_50['age'],
                       y = over_50['thalach'],
                       c = over_50['target'],
                       cmap = 'winter')
# cutomize ax1 
ax1.set( title =' heart disease and heart beat',
         ylabel= 'Heartbeat',
       )
ax1.legend(*scatter.legend_elements(), title = 'target')
ax1.axhline( y = over_50['thalach'].mean(), linestyle='--')

ax1.set_ylim([60, 200])
# add tittle to the figure 
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
plt.show()
```
